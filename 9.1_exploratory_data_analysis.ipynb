# cummulative distributive function

```python
###cdf tells us at any poin what number of data items below that poin lies

count,bin_edges = np.histogram(iris_setosa['petal_length'], bins=10 ,density=True)

pdf = count/sum(count)                  #here i am calculating fractional pdf 
print(pdf)
print(bin_edges)
### here bin is number of intervals on which you want to create cdf
## if density is kept false then it will return number of samples in the bin else if kept true returns the pdf 
## count of occurance
cdf = np.cumsum(pdf)                           #this returns cdf

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
##iris setosa cdf

count,bin_edges = np.histogram(iris_setosa['petal_length'], bins=10 ,density=True)

pdf = count/sum(count)                  #here i am calculating fractional pdf 
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)                           #this returns cdf

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


##iris versicolor cdf

count,bin_edges = np.histogram(iris_versicolor['petal_length'], bins=10 ,density=True)

pdf = count/sum(count)                  #here i am calculating fractional pdf 
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)                           #this returns cdf

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


##iris virginia cdf

count,bin_edges = np.histogram(iris_virginica['petal_length'], bins=10 ,density=True)

pdf = count/sum(count)                  #here i am calculating fractional pdf 
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)                           #this returns cdf

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

# mean variance standard-deviation

```python
print(np.mean(iris_setosa['petal_length']))                       #mean is similar to average

#mean with an outlier
print(np.mean(np.append(iris_setosa['petal_length'],50)))          ##with a single outlier mean is drastically changed
print(np.mean(iris_versicolor['petal_length']))   
print(np.mean(iris_virginica['petal_length']))   

##standard deviation
print("standard deviation")
print(np.std(iris_setosa['petal_length']))                    
print(np.std(iris_versicolor['petal_length']))   
print(np.std(iris_virginica['petal_length'])) 
```

# median,percentile,quantile,inter quantile range(IQR),MAD(median absolute difference)



```python
print(np.median(iris_setosa['petal_length']))
#median with an outlier
print(np.median(np.append(iris_setosa['petal_length'],50)))
print(np.median(iris_versicolor['petal_length']))
print(np.median(iris_virginica['petal_length']))

## quantiles of data point
print("*************************************888")
print(np.percentile(iris_setosa['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_versicolor['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_virginica['petal_length'],np.arange(0,100,25)))


## percentiles of data point
print("*************************************888")
print(np.percentile(iris_setosa['petal_length'],90))
print(np.percentile(iris_versicolor['petal_length'],90))
print(np.percentile(iris_virginica['petal_length'],90))
print("*************************************888")
print("*************************************888")

```

```python
###### MAD

from statsmodels import robust

print(robust.mad(iris_setosa['petal_length']))
print(robust.mad(iris_virginica['petal_length']))
print(robust.mad(iris_versicolor['petal_length']))
```
